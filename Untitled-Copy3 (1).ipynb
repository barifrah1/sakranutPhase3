```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
#from datetime import datetime
import sklearn

```

```python
data=pd.read_csv('ks-projects-201801_bar.csv')
copy=data
```

```python
#ignore live\cancel\other because lower density
data = data.loc[data['state'].isin(['failed','successful'])]


# drop na values
data=data.dropna()
```

```python
#data['launch_year']=pd.to_datetime(data['launched'], format="%Y/%m/%d").dt.year
#data['ratio']=data['usd_pledged_real']/data['usd_goal_real']
#feature removal
data = data.drop(['ID','goal','pledged','usd pledged',
                  'name','deadline','launched','backers','usd_pledged_real','backers_s','pledged_s'],1)
```

```python
#data['country'].value_counts(normalize=True) 
#US and gb is almost 80% of the data =>dimension reduction
data.loc[~data['country'].isin(['US','GB']),'country']='OTHER'
data.loc[~data['currency'].isin(['USD','GBP','EUR','CAD','AUD']),'currency']='OTHER'

data['duration']=10*round(data['duration']/10)
data['duration']=data['duration'].astype(int)
data=data[data['duration']>5]
```

```python

i=0
while i<=0.95:
    if i==0:
        mean=data.loc[data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.05)]['usd_goal_real'].mean()
        data.loc[(data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.05)), 'goal_level'] =int(mean)
        i+=0.05
        i=round(i,2)
        continue
    if i==0.95:
        mean=data.loc[data['usd_goal_real'] > data.usd_goal_real.quantile(i)]['usd_goal_real'].mean()
        data.loc[(data['usd_goal_real'] > data.usd_goal_real.quantile(i)), 'goal_level'] =int(mean)
        i+=0.05
        i=round(i,2)
        break
        
    if i==0.45:
        mean=data.loc[((data['usd_goal_real'] > data.usd_goal_real.quantile(i) ) & (
        data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.058) ))]['usd_goal_real'].mean()  
        data.loc[((data['usd_goal_real'] > data.usd_goal_real.quantile(i) ) & (data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.058) )), 'goal_level'] = int(mean)        
        i+=0.05
        i=round(i,2)
        continue
        
    mean=data.loc[((data['usd_goal_real'] > data.usd_goal_real.quantile(i) ) & (
            data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.05) ))]['usd_goal_real'].mean()
    
    data.loc[((data['usd_goal_real'] > data.usd_goal_real.quantile(i) ) & (
            data['usd_goal_real'] <= data.usd_goal_real.quantile(i+0.05) )), 'goal_level'] = int(mean)
    i+=0.05
    i=round(i,2)
```
